### 2번

```python
import pandas as pd
import seaborn as sns

tips = sns.load_dataset('tips')
tips.head()
```

##### 2-1번 : 팁이 7달러 이상인 데이터 추출

```python
tips[tips['tip']>=7]
```

##### 2-2번 : 여성 손님이면서 비흡연자인 데이터 찾기

```python
conditions = (tips['sex']=='Female') & (tips['smoker']=='No')
tips[conditions].head()
```

1. string형은 꼭 '' 붙이기
2. 비교할 때는 각 조건에 괄호 씌워주기

##### 2-3번 : 팁을 가장 많이 지불한 고객 찾기

```python
tips[ tips['tip']==tips['tip'].max() ]
```

##### 2-4번 : 팁을 가장 많이 지불한 고객 찾기(문제가 틀린것같음)

```python
tips.loc[ tips.sort_values(['total_bill', 'tip'], ascending=False).index[:10] ]
```

```python
# 이것도 방법이긴하나 문제의 조건은 tips[] 이 형태여야함
tips.sort_values(['total_bill', 'tip'], ascending=False).head(10)
```

##### 2-5번 : size를 기준으로 내림차, tip을 기준으로 오름차 정렬한 결과 상위 10개 데이터만 추출하기

```python
tips.sort_values(['size', 'tip'], ascending=[False,True]).head(10)
```

##### 2-6번 : total_bill과 tip 합계인 total 칼럼 생성하기

```python
tips['total'] = tips['total_bill'] + tips['tip']
tips
```

##### 2-7번 : 다음은 day별 tip의 평균 출력

문제를 풀기위해 데이터에 대한 이해가 필요하다고 느낌. 아래의 데이터는 메타데이터에 대한 이해를 위해 작성된 것

```python
tips.head(3)
```

```python
tips['day'].unique()
```

```python
pd.DataFrame(tips.groupby('day'))
```

```python
pd.DataFrame(tips.groupby('day')['tip'].mean() )
```

- 오답:   
pd.DataFrame(tips['tip'].mean().sort_values('day') )

여기서 tips['tip'].mean() 은  tip 전체의 평균값 **하나**만 구함.

- 정답: day 별로 그룹화 → 각 그룹에서 tip 평균 구하기
