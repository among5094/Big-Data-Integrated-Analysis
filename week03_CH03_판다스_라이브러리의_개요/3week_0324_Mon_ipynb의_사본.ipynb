### 3.1 판다스 데이터의 구조 이해

판다스(Pandas)는 파이썬에서 데이터를 쉽게 다룰 수 있도록 만든 라이브러리  
우리가 자주 사용하는 자료구조(data structure) 두 개있음.  

(1) 시리즈  
- 1차원 데이터 구조로, 리스트나 배열과 비슷하지만 각 요소에 인덱스(index)를 붙일 수 있음.  
- 1차원: 한 줄짜리 데이터 (열 하나)
- **인덱스(index)**와 값(value) 두 요소로 구성
- pd.Series()를 사용해 생성
- **인덱스를 숫자 대신 문자열 등으로 설정 가능**

  
(2) 데이터프레임  
- 2차원 데이터 구조로, 행(row)과 열(column)로 구성된 엑셀 표와 같은 형태의 자료.
- 여러 개의 시리즈를 합쳐서 만든 것
- pd.DataFrame()으로 생성
- 행(row): 개별 데이터
- 열(column): 변수나 속성
- 인덱스(index): 행을 구분하는 값

### loc[]

- 레이블이나 조건 표현으로 선택하는 방법: loc[]
- 인덱스와 컬럼 이름으로 접근
- 인덱스가 날짜나 문자열이라도 쓸 수 있음
- 슬라이싱할 때 마지막 값도 포함됨!





```python
# 모든 행의 여러 칼럼 선택하기: loc() 이용
df.loc[:, ['날짜', '최고기온']].head(10)
```

- loc[]는 범위를 지정할 때, 끝부분도 포함.
- 보통 슬라이싱하면 [n:m]일 때, 출력은 [n:m-1]임
- loc[]의 범위는 **[시작:끝] 범위**!

```python
df.loc[0:3, ['날짜', '최고기온']] # loc는 0번부터 3번까지 다 출력함 why?
```

```python
# 인덱스를 '날짜' 컬럼으로 지정 )
# 기존의 숫자 인덱스(0, 1, 2, ...)를 '날짜' 컬럼 값으로 바꾼경우
df.index=df['날짜']

# '2022-06-05'라는 인덱스 값(=행)을 기준으로 '최저기온', '최고기온' 선택
# 특정 날짜에 해당하는 컬럼 선택하기
df.loc['2022-06-05', ['최저기온', '최고기온']]
```

```python
# 평균기온이 25도 이상인 모든 행 전체를 선택
df.loc[df['평균기온']>=25]
```

```python
df.loc[df['평균기온']>=25,  ['최저기온', '최고기온']]
```

##### 3.4.3 불 인덱싱

### 3.5 데이터 가공 및 그룹핑

```python
filename='/content/drive/MyDrive/data/music.csv'
df=pd.read_csv(filename)
df.head()
```

##### 3.5.1 칼럼 삭제

drop() 메서드
- 열을 삭제하는 메서드
- pop과는 다르게 데이터프레임 원본이 변경되지 않는다.
- inplace 옵션은 원본을 변경할지말지의 여부 -> True이면 원본 변경됨
- axis=1은 삭제 대상이 열, axis=0은 삭제 대상이 행


```python
# 데이터 분석에 필요없는 agency(기획사) 삭제
df.drop(columns=['agency'],axis=1, inplace=True)
```

```python
df.columns # 열 확인하기
```

##### 3.5.2 칼럼 이름 변경


```python
# 모든 컬럼 이름 변경하기
# 전체 변수 이름을 재설정
# 변수명을 차례로 재설정한다.
df.columns=['제목', '이름', '인원수', '구분', '발매일', '장르', '유형']
df.head()
```

```python
df.dtypes # 데이터타입을 보면 발매일은 숫자로 보이지만 사실 문자열임
```

```python
df['발매일'] = pd.to_datetime(  df['발매일'], errors='coerce') # 발매일을 datetime64로 변경
```

```python
df.dtypes # 다시 확인하면 시간 형태로 바뀜
```

##### 3.5.3 칼럼 생성

```python
# 새로운 컬럼 생성하기: 년도, 월, 일
df['년도']=df['발매일'].dt.year # 발매일에서 연도만 추출해서 년도 컬럼으로 저장
df['월']=df['발매일'].dt.month # 발매일에서 월(month)만 추출
df['일']=df['발매일'].dt.day # 발매일에서 일(day)만 추출
df.head()
```

##### 3.5.4 데이터그룹핑

```python
# 한 열을 기준으로 그룹화하기
# '인원수' 값이 존재하는 행의 개수
newdata = df.groupby('장르').count() # '장르' 컬럼을 기준으로 그룹화하고, NaN이 아닌 값의 개수를 모든 열에 대해 계산
pd.DataFrame(newdata['인원수'])
# newdata['인원수'] 랑 똑같음.
# 굳이 pd.DataFrame()으로 안 감싸도됨. 그냥 표 형태로 보기 위해서 감싸는 것일뿐
```

```python
# 여러 열을 기준으로 그룹화하기
newdata = df.groupby(['구분', '장르']).sum()
pd.DataFrame(newdata['인원수'])
```

결측데이터 대체: 최빈값으로 채우기

```python
# 최빈값 찾기
df['Embarked'].mode()
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
# 결과를 보면 Embarked가 0이다
```

```python
# 평균값으로 대체
avg=df['Age'].mean() # 나이의 평균값을 구해
df['Age']=df['Age'].fillna(avg) # 평균값으로 대체
```
