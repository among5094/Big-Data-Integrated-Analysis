### 3.1 판다스 데이터의 구조 이해

(1) 시리즈  
- 1차원 데이터 구조로, 리스트나 배열과 비슷하지만 각 요소에 인덱스(index)를 붙일 수 있음.  
- 1차원: 한 줄짜리 데이터 (열 하나)
- **인덱스(index)**와 값(value) 두 요소로 구성
- pd.Series()를 사용해 생성
- **인덱스를 숫자 대신 문자열 등으로 설정 가능**

  
(2) 데이터프레임  
- 2차원 데이터 구조로, 행(row)과 열(column)로 구성된 엑셀 표와 같은 형태의 자료.
- 여러 개의 시리즈를 합쳐서 만든 것
- pd.DataFrame()으로 생성
- 행(row): 개별 데이터
- 열(column): 변수나 속성
- 인덱스(index): 행을 구분하는 값

### 3.5 데이터 가공 및 그룹핑

```python
filename='/content/drive/MyDrive/data/music.csv'
df=pd.read_csv(filename)
df.head()
```

```python
# 뎅터 분석에 필요없는 agency(기획사) 삭제
df.drop(columns=['agency'],axis=1, inplace=True)
```

##### 3.5.(2) 칼럼 이름 변경


```python
# 모든 컬럼 이름 변경하기
df.columns=['제목', '이름', '인원수', '구분', '발매일', '장르', '유형']
df.head()
```

```python
df.dtypes # 데이터타입을 보면 발매일은 숫자로 보이지만 사실 문자열임
```

```python
df['발매일'] = pd.to_datetime(  df['발매일'], errors='coerce') # 발매일을 datetime64로 변경
```

```python
df.dtypes # 다시 확인하면 시간 형태로 바뀜
```

```python
# 새로운 컬럼 생성하기: 년도, 월, 일
df['년도']=df['발매일'].dt.year
df['월']=df['발매일'].dt.month
df['일']=df['발매일'].dt.day
df.head()
```

```python
# 한 열을 기준으로 그룹화하기
newdata = df.groupby('장르').count()
pd.DataFrame(newdata['인원수'])
```

```python
# 여러 열을 기준으로 그룹화하기
newdata = df.groupby(['구분', '장르']).count()
pd.DataFrame(newdata['인원수'])
```

### 3.6 결측치 처리하기

```python
# 데이터 불러오기
filename='/content/drive/MyDrive/data/train.csv'
df=pd.read_csv(filename)
df.head()
```

```python
# 결츨데이터 확인: 탑승항구(Embarked) 결측 데이터
df[df['Embarked'].isnull()]
```

```python
# 결츨데이터 확인: 객실번호(Cabin) 결측 데이터
df[df['Cabin'].isnull()].head(20)
```

```python
# 컬럼별 결측 데이터 개수 확인
df.isnull().sum()
# df.isna().sum() 기능이 완전히 똑같음!

# 0은 누락데이터가 없는 것이고, age는 177개가 없고, Cabin은 687개가 없다. 탑승항구는 2개 없음
```

##### 결측데이터 삭제

```python
# 데이터 삭제 - Cabin은 누락데이터가 너무 많아서 그냥 삭제하기로 결정
df.drop('Cabin', axis=1, inplace=True)
```

```python
df.columns
```

결측데이터 대체: 최빈값으로 채우기

```python
# 최빈값 찾기
df['Embarked'].mode()
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
# 결과를 보면 Embarked가 0이다
```

```python
# 평균값으로 대체
avg=df['Age'].mean() # 나이의 평균값을 구해
df['Age']=df['Age'].fillna(avg) # 평균값으로 대체
```

```python
df.isnull().sum() # 결과를 보면 age가 0으로 되어있음
```
