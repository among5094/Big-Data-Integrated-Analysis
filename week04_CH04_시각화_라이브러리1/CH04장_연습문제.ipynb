### 1번 데이터 시각화의 필요성에 대해 작성하시오.

데이터 시각화는 복잡한 데이터를 시각적으로 표현하여 더 쉽게 이해하고, 인사이트를 도출할 수 있도록 도와주는 중요한 과정입니다.
- **의미 전달의 용이성**: 수치 데이터보다 시각적 자료가 더 직관적으로 의미를 전달함.
- **데이터 요약**: 많은 양의 데이터를 한눈에 요약할 수 있음.
패턴과 이상값 파악: 시각화를 통해 숨겨진 패턴, 트렌드, 이상값을 쉽게 발견할 수 있음.
- **정확한 분석**: 분석 결과를 시각화하면 분석의 타당성을 시각적으로 검증할 수 있음.
- **커뮤니케이션 도구**: 동료나 의사결정자에게 결과를 효과적으로 전달하는 데 활용됨.


### 2번 시각화 라이브러리 종류를 나열하고 차이점을 작성하시오.
- **Matplotlib**: 가장 기본적인 시각화 라이브러리. 선, 막대, 히스토그램 등 다양한 그래프 표현 가능. 커스터마이징 용이함.
- **Seaborn**: Matplotlib 기반으로 더 세련된 시각화 가능. 통계 시각화에 강점. 색상, 스타일이 기본적으로 예쁨.
- **Plotly**: 대화형 그래프 지원. 줌/툴팁 기능 제공. 웹 기반 시각화에 적합.

### 4번 문제: 아이리스 붓꽃 데이터를 이용하여 다음 선 그래프와 같이 출력되도록 빈칸에 알맞은 코드를 완성하시오.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 준비
data = pd.read_csv('/content/Drive/MyDrive/data/04_data2.csv')

# 데이터 옮기기
PetalLength = data['PetalLength']
PetalWidth = data['PetalWidth']
SepaLength = data['SepalLength']
SepaWidth = data['SepalWidth']

plt.figure(figsize=(15,4)) # 도화지 크기 설정

# < 빈칸 채우기 >
plt.plot(PetalLength, color='r', marker='o', label='Petal Length')
plt.plot(PetalWidth, color='g', marker='o', label='Petal Width')
plt.plot(SepaLength, color='b', marker='o', label='Sepal Length')
plt.plot(SepaWidth, color='y', marker='o', label='Sepal Width')


plt.xlabel('Iris')
plt.ylabel('Value')
plt.title('Iris Line Graph')
plt.legend()
plt.grid() # 눈금
plt.show()

```

### 5번 아이리스 붓꽃 데이터를 이용하여 다음 막대 그래프와 같이 출력되도록 빈칸에 알맞은 코드를 완성하시오.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 준비
data = pd.read_csv('/content/Drive/MyDrive/data/04_data2.csv')

# 데이터 옮기기
PetalLength = data['PetalLength'].mean()
PetalWidth = data['PetalWidth'].mean()
SepaLength = data['SepalLength'].mean()
SepaWidth = data['SepalWidth'].mean()

plt.figure(figsize=(10,5)) # 도화지 크기 설정

# < 빈칸 채우기 >
plt.bar('PetalLength', PetalLength)
plt.bar('PetalWidth', PetalWidth)
plt.bar('SepaLength',SepaLength)
plt.bar('SepaWidth',SepaWidth)

plt.xlabel('Iris')
plt.ylabel('Value')
plt.title('Iris bar Graph')
plt.grid() # 눈금
plt.show()
```

###  6번 아이리스 붓꽃 데이터를 이용하여 다음 산점도 그래프와 같이 출력되도록 빈칸에 알맞은 코드를 완성하시오.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 준비
data = pd.read_csv('/content/Drive/MyDrive/data/04_data2.csv')
SepalLength = data['SepalLength']
SepalWidth = data['SepalWidth']

plt.figure(figsize=(10,4))

# 빈칸 채우기
plt.scatter(SepalLength, SepalWidth)

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Iris Sepal Scatter Graph')
plt.grid()
plt.show()
```

###  **7번 아이리스 붓꽃 데이터를 이용하여 다음 히스토그램과 같이 출력되도록 빈칸에 알맞은 코드를 완성하시오.**

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/Drive/MyDrive/data/04_data2.csv')

virginica = data.loc[data.Species == 'virginica', ['Species', 'SepalLength']]

plt.figure(figsize=(10,4))

# 빈칸 채우기
plt.hist(virginica['SepalLength'], bins=7, label='Sepal Length')

plt.xlabel('Sepal')
plt.ylabel('Length')
plt.title('Iris Virginica Sepal Histogram')
plt.legend()
plt.grid()
plt.show()

```

### 8번 **미세먼지 등급별 일수 데이터 또는 아이리스 붓꽃 데이터를 이용하여 히스토그램, 상자수염 그래프를 작성하시오.**
