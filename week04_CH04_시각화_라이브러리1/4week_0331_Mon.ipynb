```python
# '2015년 미세먼지 매우 나쁨' 세로 막대 그래프
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 불러오기
data = pd.read_excel('/content/Drive/MyDrive/data/04_data1.xlsx', index_col='area')
data2015_verybad = data['2015_verybad']

plt.figure(figsize=(20, 4))
plt.bar(data2015_verybad.index, data2015_verybad, color='g') # 막대그래프 그리기

plt.xlabel('Area')
plt.ylabel('Micrometer')

plt.title('2015 Fine Dust(pm10) Very Bad Bar Graph')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# x축의 기준이 될 도시 인덱스 0~6 생성 ([0 1 2 3 4 5 6])
index = np.arange(7)

# 넓은 그래프
plt.figure(figsize=(20, 4))

# 범위가 '서울'에서 '울산'까지, 2015_good에서 2020_good까지 읽음
data6 = data.loc['Seoul':'Ulsan', '2015_good':'2020_good']

#
for year in range(2015, 2021):
    year_col = str(year) + '_good' # 컬럼 이름 만들기
    chartdata = data6[year_col]  # 해당 연도에 해당하는 데이터만 추출
    plt.bar(index, chartdata, width=0.15, label=year_col) # 막대그래프 그리기
    index = index + 0.15 # 이게 없으면 막대들이 겹쳐서 그룹별 하나의 막대로 보임

plt.xlabel('Area')
plt.ylabel('Micrometer')

#  라벨이 막대 묶음의 중앙에 오도록 -0.53 해주기
plt.xticks(index - 0.53, ['Seoul', 'Busan', 'Daegu', 'Incheon', 'Gwangju', 'Daejeon', 'Ulsan']) # 중앙에 라벨 붙이기

plt.title('2015~2020 Fine Dust(pm10) Good Group Bar Graph')
plt.legend()
plt.show()
```

##### 3) 그룹 누적 가로 막대 그래프 그리기

```python
import numpy as np
index = np.arange(7)

plt.figure(figsize=(10, 4))

data6 = data.loc['Seoul':'Ulsan', '2017_common':'2020_common']
for year in range(2017, 2021):
    year = str(year) + '_common'
    chartdata = data6[year]
    plt.barh(index, chartdata, label=year)

plt.xlim(120, 310)
plt.ylabel('Area')
plt.xlabel('Micrometer')
plt.yticks(index, ['Seoul', 'Busan', 'Daegu', 'Incheon', 'Gwangju', 'Daejeon', 'Ulsan'])
plt.title('2017~2020 Fine Dust(pm10) Common Group Barh Graph')
plt.legend()
plt.show()
```

### 4.2(4) 산점도 그래프

```python
import pandas as pd
import matplotlib.pyplot as plt

# 아이리스 붓꽃 데이터 읽어오기
data = pd.read_csv('/content/MyDrive/MyDrive/data/04_data2.csv')
data
```

```python
# 산점도 그래프 그리기: scatter()로 표현
PetalLength = data['PetalLength']
PetalWidth = data['PetalWidth']

plt.figure(figsize=(10,4))
plt.scatter(PetalLength, PetalWidth)

plt.xlabel('PetalLength')
plt.ylabel('PetalWidth')
plt.title('Iris Petal Scatter Graph')
plt.grid()
plt.show()
```

```python
# 그룹 산점도 그래프 그리기
PetalLength = data['PetalLength']
PetalWidth = data['PetalWidth']

SepalLength = data['SepalLength']
SepalWidth = data['SepalWidth']

plt.figure(figsize=(10,4))
plt.scatter(PetalLength, PetalWidth, color='r', edgecolor='w',
            label='PetalLength*PetalWidth')
plt.scatter(SepalLength, SepalWidth, color='b', edgecolor='w',
            label='SepalLength*SepalWidth')

plt.xlabel('PetalLength')
plt.ylabel('PetalWidth')
plt.title('Iris Petal Scatter Graph')
plt.grid()
plt.show()

```

### 4.2(5) 히스토그램

```python
setosa = data.loc[data.Species=='setosa', ['Species','PetalLength' ]]

plt.figure(figsize=(10,6))
plt.hist(setosa['PetalLength'], bins=8, label='Petal Length')
# 8개의 칸으로 구성된 거 그리겠다는 뜻인데 실제 이미지는 8칸이 좀 오버라 잘림

plt.xlabel('Petal')
plt.ylabel('Length')
plt.title('Iris Setosa Petal Histogram')
plt.legend()
plt.grid()
plt.show()
```

```python
# 아이리스 붓꽃 그룹 히스토그램


versicolor_PetalLength = data.loc[data.Species == 'versicolor', ['Species', 'PetalLength']]
virginica_PetalLength = data.loc[data.Species == 'virginica', ['Species', 'PetalLength']]

plt.figure(figsize=(10, 6))
plt.hist(versicolor_PetalLength['PetalLength'], bins=5, alpha=0.5, label='versicolor')
plt.hist(virginica_PetalLength['PetalLength'], bins=5, alpha=0.5, label='virginica')
# alpha : 투명도 / 1은 불투명, 0은 투명

plt.xlabel('Petal')
plt.ylabel('Length')
plt.title('Iris Petal Group Histogram')
plt.legend()
plt.grid()
plt.show()

```

### 4.2(6) 박스플롯

```python
# 아이리스 붓꽃 박스플롯 그래프
SepalLength = data['SepalLength']
SepalWidth = data['SepalWidth']

plt.figure(figsize=(8, 5))
plt.boxplot([SepalLength, SepalWidth], labels=['Sepal Length', 'Sepal Width'])

plt.xlabel('Sepal')
plt.ylabel('Value')
plt.title('Iris Sepal Box Plot')
plt.show()
```

```python
# 아이리스 붓꽃 가로 박스플롯 그래프

plt.figure(figsize=(8, 5))
plt.boxplot([SepalLength, SepalWidth], labels=['Sepal Length', 'Sepal Width'], vert=False)

plt.ylabel('Sepal')
plt.xlabel('Value')
plt.title('Iris Sepal Horizontal Box Plot')
plt.show()
```
